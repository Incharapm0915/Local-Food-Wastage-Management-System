```python
# EDA and Analysis for Local Food Wastage Management System
# ========================================================

# Cell 1: Import Libraries and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import sys
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Add project root to path
project_root = Path.cwd().parent
sys.path.append(str(project_root))

from src.analysis.sql_queries import FoodWastageAnalyzer

# Set up plotting
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)

print("üìä Food Wastage Management System - EDA & Analysis")
print("="*60)

# Initialize analyzer
analyzer = FoodWastageAnalyzer()

# Test connection
if analyzer.db.get_connection() is None:
    print("‚ùå Database connection failed!")
    exit()
else:
    print("‚úÖ Database connected successfully!")

# Cell 2: Execute All Queries and Store Results
print("\nüîç Executing all analytical queries...")
results = analyzer.run_all_queries(save_results=False)

print(f"\nüìä Query Results Summary:")
for query_name, df in results.items():
    print(f"   ‚Ä¢ {query_name}: {len(df)} rows")

# Cell 3: Provider Analysis Visualizations
print("\nüè¢ PROVIDER ANALYSIS")
print("="*40)

# Query 2: Provider Types Analysis
provider_types = results.get("Query 2: Top Provider Types")
if provider_types is not None:
    print("üìà Provider Types Contribution:")
    print(provider_types.to_string(index=False))
    
    # Visualization 1: Provider Types Pie Chart
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Pie chart of quantity contribution
    ax1.pie(provider_types['total_quantity_contributed'], 
            labels=provider_types['provider_type'],
            autopct='%1.1f%%', startangle=90)
    ax1.set_title('Food Quantity Contribution by Provider Type')
    
    # Bar chart of number of providers
    ax2.bar(provider_types['provider_type'], provider_types['number_of_providers'])
    ax2.set_title('Number of Providers by Type')
    ax2.tick_params(axis='x', rotation=45)
    
    # Bar chart of average quantity per listing
    ax3.bar(provider_types['provider_type'], provider_types['avg_quantity_per_listing'])
    ax3.set_title('Average Quantity per Listing by Provider Type')
    ax3.tick_params(axis='x', rotation=45)
    
    # Combined chart: Providers vs Total Listings
    ax4.scatter(provider_types['number_of_providers'], 
               provider_types['total_food_listings'],
               s=provider_types['total_quantity_contributed']/100,
               alpha=0.7)
    ax4.set_xlabel('Number of Providers')
    ax4.set_ylabel('Total Food Listings')
    ax4.set_title('Providers vs Listings (bubble size = quantity)')
    
    plt.tight_layout()
    plt.show()

# Query 1: Geographic Distribution
geo_distribution = results.get("Query 1: Providers & Receivers by City")
if geo_distribution is not None:
    print(f"\nüó∫Ô∏è  Geographic Distribution (Top 10 cities):")
    print(geo_distribution.head(10).to_string(index=False))
    
    # Visualization 2: Geographic Distribution
    top_cities = geo_distribution.head(15)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Stacked bar chart
    x = range(len(top_cities))
    ax1.bar(x, top_cities['providers'], label='Providers', alpha=0.8)
    ax1.bar(x, top_cities['receivers'], bottom=top_cities['providers'], 
            label='Receivers', alpha=0.8)
    ax1.set_xlabel('Cities')
    ax1.set_ylabel('Count')
    ax1.set_title('Providers and Receivers by City (Top 15)')
    ax1.set_xticks(x)
    ax1.set_xticklabels(top_cities['city'], rotation=45, ha='right')
    ax1.legend()
    
    # Scatter plot: Providers vs Receivers
    ax2.scatter(top_cities['providers'], top_cities['receivers'], 
               s=top_cities['total_entities']*3, alpha=0.6)
    ax2.set_xlabel('Number of Providers')
    ax2.set_ylabel('Number of Receivers')
    ax2.set_title('Providers vs Receivers by City')
    
    # Add city labels to points
    for _, row in top_cities.head(8).iterrows():
        ax2.annotate(row['city'], (row['providers'], row['receivers']), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    plt.tight_layout()
    plt.show()

# Cell 4: Food Availability Analysis
print("\nüçΩÔ∏è  FOOD AVAILABILITY ANALYSIS")
print("="*40)

# Query 5: Total Food Available
total_food = results.get("Query 5: Total Food Available")
if total_food is not None:
    print("üìä Overall Food Availability:")
    print(total_food.to_string(index=False))

# Query 7: Food Types Analysis
food_types = results.get("Query 7: Common Food Types")
if food_types is not None:
    print(f"\nü•ó Food Types Analysis:")
    print(food_types.to_string(index=False))
    
    # Visualization 3: Food Types Analysis
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Pie chart of quantity by food type
    ax1.pie(food_types['total_quantity'], labels=food_types['food_type'],
            autopct='%1.1f%%', startangle=90)
    ax1.set_title('Food Quantity Distribution by Type')
    
    # Bar chart of listings by food type
    ax2.bar(food_types['food_type'], food_types['number_of_listings'])
    ax2.set_title('Number of Listings by Food Type')
    ax2.tick_params(axis='x', rotation=45)
    
    # Average quantity per listing
    ax3.bar(food_types['food_type'], food_types['avg_quantity_per_listing'])
    ax3.set_title('Average Quantity per Listing by Food Type')
    ax3.tick_params(axis='x', rotation=45)
    
    # Providers offering each food type
    ax4.bar(food_types['food_type'], food_types['providers_offering'])
    ax4.set_title('Number of Providers by Food Type')
    ax4.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

# Query 12: Meal Type Popularity
meal_types = results.get("Query 12: Meal Type Popularity")
if meal_types is not None:
    print(f"\nüçΩÔ∏è  Meal Type Analysis:")
    print(meal_types.to_string(index=False))
    
    # Visualization 4: Meal Type Analysis
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
    
    # Total quantity claimed by meal type
    ax1.bar(meal_types['meal_type'], meal_types['total_quantity_claimed'])
    ax1.set_title('Total Quantity Claimed by Meal Type')
    ax1.tick_params(axis='x', rotation=45)
    ax1.set_ylabel('Quantity Claimed')
    
    # Success rate by meal type
    ax2.bar(meal_types['meal_type'], meal_types['success_rate_percentage'])
    ax2.set_title('Claim Success Rate by Meal Type')
    ax2.tick_params(axis='x', rotation=45)
    ax2.set_ylabel('Success Rate (%)')
    
    # Utilization rate by meal type
    ax3.bar(meal_types['meal_type'], meal_types['utilization_rate_percentage'])
    ax3.set_title('Food Utilization Rate by Meal Type')
    ax3.tick_params(axis='x', rotation=45)
    ax3.set_ylabel('Utilization Rate (%)')
    
    plt.tight_layout()
    plt.show()

# Cell 5: Claims and Distribution Analysis
print("\nüìã CLAIMS & DISTRIBUTION ANALYSIS")
print("="*40)

# Query 10: Claim Status Distribution
claim_status = results.get("Query 10: Claim Status Distribution")
if claim_status is not None:
    print("üìä Claim Status Distribution:")
    # Exclude TOTAL row for cleaner display
    status_data = claim_status[claim_status['status'] != 'TOTAL']
    print(status_data.to_string(index=False))
    
    # Visualization 5: Claim Status Analysis
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Pie chart of claim status
    ax1.pie(status_data['claim_count'], labels=status_data['status'],
            autopct='%1.1f%%', startangle=90,
            colors=['#2ecc71', '#f39c12', '#e74c3c'])  # Green, Orange, Red
    ax1.set_title('Distribution of Claim Status')
    
    # Bar chart with percentages
    bars = ax2.bar(status_data['status'], status_data['claim_count'],
                   color=['#2ecc71', '#f39c12', '#e74c3c'])
    ax2.set_title('Claim Counts by Status')
    ax2.set_ylabel('Number of Claims')
    
    # Add percentage labels on bars
    for bar, percentage in zip(bars, status_data['percentage']):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                f'{percentage}%', ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

# Query 4: Top Receivers Analysis
top_receivers = results.get("Query 4: Top Food Claimers")
if top_receivers is not None:
    print(f"\nüéØ Top 10 Food Receivers:")
    print(top_receivers.head(10)[['receiver_name', 'receiver_type', 'city', 
                                  'total_claims', 'success_rate_percentage', 
                                  'total_food_received']].to_string(index=False))
    
    # Visualization 6: Top Receivers Analysis
    top_10_receivers = top_receivers.head(10)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Total food received by top receivers
    ax1.barh(range(len(top_10_receivers)), top_10_receivers['total_food_received'])
    ax1.set_yticks(range(len(top_10_receivers)))
    ax1.set_yticklabels(top_10_receivers['receiver_name'], fontsize=8)
    ax1.set_xlabel('Total Food Received')
    ax1.set_title('Top 10 Receivers by Food Quantity')
    
    # Success rate vs total claims scatter
    ax2.scatter(top_10_receivers['total_claims'], 
               top_10_receivers['success_rate_percentage'],
               s=top_10_receivers['total_food_received']/10, alpha=0.6)
    ax2.set_xlabel('Total Claims')
    ax2.set_ylabel('Success Rate (%)')
    ax2.set_title('Claims vs Success Rate (bubble size = food received)')
    
    plt.tight_layout()
    plt.show()

# Cell 6: Provider Performance Analysis
print("\nüè¢ PROVIDER PERFORMANCE ANALYSIS")
print("="*40)

# Query 9: Successful Providers
successful_providers = results.get("Query 9: Successful Providers")
if successful_providers is not None:
    print(f"\nüèÜ Top 10 Successful Providers:")
    print(successful_providers.head(10)[['provider_name', 'provider_type', 'city',
                                        'successful_claims', 'success_rate_percentage',
                                        'total_food_distributed']].to_string(index=False))
    
    # Visualization 7: Provider Performance
    top_providers = successful_providers.head(12)
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Top providers by successful claims
    ax1.barh(range(len(top_providers)), top_providers['successful_claims'])
    ax1.set_yticks(range(len(top_providers)))
    ax1.set_yticklabels(top_providers['provider_name'], fontsize=8)
    ax1.set_xlabel('Successful Claims')
    ax1.set_title('Top Providers by Successful Claims')
    
    # Success rate distribution
    ax2.hist(successful_providers['success_rate_percentage'], bins=20, alpha=0.7)
    ax2.set_xlabel('Success Rate (%)')
    ax2.set_ylabel('Number of Providers')
    ax2.set_title('Distribution of Provider Success Rates')
    
    # Provider type performance
    provider_type_perf = successful_providers.groupby('provider_type').agg({
        'successful_claims': 'sum',
        'total_food_distributed': 'sum',
        'success_rate_percentage': 'mean'
    }).reset_index()
    
    ax3.bar(provider_type_perf['provider_type'], 
           provider_type_perf['total_food_distributed'])
    ax3.set_title('Total Food Distributed by Provider Type')
    ax3.tick_params(axis='x', rotation=45)
    ax3.set_ylabel('Food Distributed')
    
    # Average success rate by provider type
    ax4.bar(provider_type_perf['provider_type'], 
           provider_type_perf['success_rate_percentage'])
    ax4.set_title('Average Success Rate by Provider Type')
    ax4.tick_params(axis='x', rotation=45)
    ax4.set_ylabel('Average Success Rate (%)')
    
    plt.tight_layout()
    plt.show()

# Cell 7: Geographic Efficiency Analysis
print("\nüó∫Ô∏è  GEOGRAPHIC EFFICIENCY ANALYSIS")
print("="*40)

# Query 14: Geographic Food Distribution
geo_efficiency = results.get("Query 14: Geographic Distribution")
if geo_efficiency is not None:
    print(f"\nüìç Top 10 Cities by Food Distribution:")
    print(geo_efficiency.head(10)[['city', 'total_food_available', 'food_distributed',
                                  'claim_success_rate', 'food_utilization_rate']].to_string(index=False))
    
    # Visualization 8: Geographic Efficiency
    top_cities_eff = geo_efficiency.head(12)
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Food available vs distributed
    ax1.scatter(top_cities_eff['total_food_available'], 
               top_cities_eff['food_distributed'],
               s=top_cities_eff['total_providers']*20, alpha=0.6)
    ax1.set_xlabel('Total Food Available')
    ax1.set_ylabel('Food Distributed')
    ax1.set_title('Food Available vs Distributed by City')
    ax1.plot([0, top_cities_eff['total_food_available'].max()], 
             [0, top_cities_eff['total_food_available'].max()], 'r--', alpha=0.5)
    
    # Utilization rate by city
    ax2.bar(range(len(top_cities_eff)), top_cities_eff['food_utilization_rate'])
    ax2.set_xlabel('Cities')
    ax2.set_ylabel('Utilization Rate (%)')
    ax2.set_title('Food Utilization Rate by City')
    ax2.set_xticks(range(len(top_cities_eff)))
    ax2.set_xticklabels(top_cities_eff['city'], rotation=45, ha='right')
    
    # Success rate vs utilization rate
    ax3.scatter(top_cities_eff['claim_success_rate'], 
               top_cities_eff['food_utilization_rate'],
               s=top_cities_eff['total_claims']/2, alpha=0.6)
    ax3.set_xlabel('Claim Success Rate (%)')
    ax3.set_ylabel('Food Utilization Rate (%)')
    ax3.set_title('Success Rate vs Utilization Rate')
    
    # Providers vs receivers efficiency
    ax4.scatter(top_cities_eff['total_providers'], 
               top_cities_eff['unique_receivers_served'],
               s=top_cities_eff['food_distributed']/50, alpha=0.6)
    ax4.set_xlabel('Number of Providers')
    ax4.set_ylabel('Unique Receivers Served')
    ax4.set_title('Provider-Receiver Network Efficiency')
    
    plt.tight_layout()
    plt.show()

# Cell 8: System Performance Metrics
print("\nüìä SYSTEM PERFORMANCE METRICS")
print("="*40)

# Query 15: Comprehensive System Metrics
system_metrics = results.get("Query 15: System Metrics")
if system_metrics is not None:
    print("üéØ Key System Metrics:")
    print(system_metrics.to_string(index=False))
    
    # Create a dashboard-style visualization
    fig = make_subplots(
        rows=3, cols=2,
        subplot_titles=('System Overview', 'Food Distribution', 
                       'Claims Performance', 'Efficiency Metrics',
                       'Geographic Coverage', 'User Engagement'),
        specs=[[{"type": "indicator"}, {"type": "indicator"}],
               [{"type": "bar"}, {"type": "pie"}],
               [{"type": "scatter"}, {"type": "bar"}]]
    )
    
    # Add indicators for key metrics
    total_providers = system_metrics[system_metrics['metric_name'] == 'Total Active Providers']['metric_value'].iloc[0]
    total_receivers = system_metrics[system_metrics['metric_name'] == 'Total Active Receivers']['metric_value'].iloc[0]
    
    fig.add_trace(go.Indicator(
        mode = "gauge+number",
        value = total_providers,
        title = {'text': "Active Providers"},
        gauge = {'axis': {'range': [None, total_providers*1.2]},
                'bar': {'color': "darkblue"},
                'steps': [{'range': [0, total_providers*0.5], 'color': "lightgray"},
                         {'range': [total_providers*0.5, total_providers], 'color': "gray"}],
                'threshold': {'line': {'color': "red", 'width': 4},
                            'thickness': 0.75, 'value': total_providers*0.9}}),
        row=1, col=1)
    
    fig.add_trace(go.Indicator(
        mode = "gauge+number",
        value = total_receivers,
        title = {'text': "Active Receivers"},
        gauge = {'axis': {'range': [None, total_receivers*1.2]},
                'bar': {'color': "darkgreen"},
                'steps': [{'range': [0, total_receivers*0.5], 'color': "lightgray"},
                         {'range': [total_receivers*0.5, total_receivers], 'color': "gray"}],
                'threshold': {'line': {'color': "red", 'width': 4},
                            'thickness': 0.75, 'value': total_receivers*0.9}}),
        row=1, col=2)
    
    fig.update_layout(height=800, showlegend=False, 
                     title_text="Food Wastage Management System - Performance Dashboard")
    fig.show()

# Cell 9: Receiver Type Analysis
print("\nüë• RECEIVER TYPE ANALYSIS")
print("="*40)

# Query 11: Average Food per Receiver
receiver_analysis = results.get("Query 11: Average Food per Receiver")
if receiver_analysis is not None:
    print("üìä Receiver Type Performance:")
    print(receiver_analysis.to_string(index=False))
    
    # Visualization 9: Receiver Type Analysis
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Average food per receiver by type
    ax1.bar(receiver_analysis['receiver_type'], 
           receiver_analysis['avg_food_per_receiver'])
    ax1.set_title('Average Food per Receiver by Type')
    ax1.tick_params(axis='x', rotation=45)
    ax1.set_ylabel('Average Food Quantity')
    
    # Total food received by receiver type
    ax2.bar(receiver_analysis['receiver_type'], 
           receiver_analysis['total_food_received'])
    ax2.set_title('Total Food Received by Receiver Type')
    ax2.tick_params(axis='x', rotation=45)
    ax2.set_ylabel('Total Food Quantity')
    
    # Number of receivers by type
    ax3.pie(receiver_analysis['total_receivers'], 
           labels=receiver_analysis['receiver_type'],
           autopct='%1.1f%%', startangle=90)
    ax3.set_title('Distribution of Receiver Types')
    
    # Average successful claims per receiver
    ax4.bar(receiver_analysis['receiver_type'], 
           receiver_analysis['avg_successful_claims_per_receiver'])
    ax4.set_title('Avg Successful Claims per Receiver by Type')
    ax4.tick_params(axis='x', rotation=45)
    ax4.set_ylabel('Average Successful Claims')
    
    plt.tight_layout()
    plt.show()

# Cell 10: Food Item Analysis
print("\nüçΩÔ∏è  FOOD ITEM ANALYSIS")
print("="*40)

# Query 8: Claims per Food Item
food_item_analysis = results.get("Query 8: Claims per Food Item")
if food_item_analysis is not None:
    print(f"üìä Top 10 Most Claimed Food Items:")
    top_claimed_foods = food_item_analysis.head(10)
    print(top_claimed_foods[['food_name', 'food_type', 'meal_type', 'location',
                            'total_claims', 'claim_success_rate']].to_string(index=False))
    
    # Visualization 10: Food Item Analysis
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
    
    # Top food items by claims
    ax1.barh(range(len(top_claimed_foods)), top_claimed_foods['total_claims'])
    ax1.set_yticks(range(len(top_claimed_foods)))
    ax1.set_yticklabels(top_claimed_foods['food_name'], fontsize=8)
    ax1.set_xlabel('Total Claims')
    ax1.set_title('Top 10 Food Items by Claims')
    
    # Success rate distribution
    ax2.hist(food_item_analysis['claim_success_rate'].dropna(), bins=20, alpha=0.7)
    ax2.set_xlabel('Claim Success Rate (%)')
    ax2.set_ylabel('Number of Food Items')
    ax2.set_title('Distribution of Food Item Success Rates')
    
    # Claims vs available quantity
    ax3.scatter(food_item_analysis['available_quantity'], 
               food_item_analysis['total_claims'],
               alpha=0.6)
    ax3.set_xlabel('Available Quantity')
    ax3.set_ylabel('Total Claims')
    ax3.set_title('Available Quantity vs Claims')
    
    # Success rate vs total claims
    valid_data = food_item_analysis.dropna(subset=['claim_success_rate'])
    ax4.scatter(valid_data['total_claims'], 
               valid_data['claim_success_rate'],
               alpha=0.6)
    ax4.set_xlabel('Total Claims')
    ax4.set_ylabel('Success Rate (%)')
    ax4.set_title('Claims vs Success Rate')
    
    plt.tight_layout()
    plt.show()

# Cell 11: Summary Insights and Recommendations
print("\nüéØ KEY INSIGHTS & RECOMMENDATIONS")
print("="*50)

def generate_insights():
    """Generate key insights from the analysis"""
    
    insights = []
    
    # Provider insights
    if "Query 2: Top Provider Types" in results:
        provider_data = results["Query 2: Top Provider Types"]
        top_provider_type = provider_data.iloc[0]
        insights.append(f"üè¢ {top_provider_type['provider_type']} providers contribute the most food ({top_provider_type['percentage_of_total_quantity']:.1f}% of total)")
    
    # Geographic insights
    if "Query 14: Geographic Distribution" in results:
        geo_data = results["Query 14: Geographic Distribution"]
        top_city = geo_data.iloc[0]
        insights.append(f"üó∫Ô∏è  {top_city['city']} is the most active city with {top_city['food_distributed']:,.0f} units of food distributed")
    
    # Efficiency insights
    if "Query 10: Claim Status Distribution" in results:
        status_data = results["Query 10: Claim Status Distribution"]
        completed_row = status_data[status_data['status'] == 'Completed']
        if not completed_row.empty:
            success_rate = completed_row.iloc[0]['percentage']
            insights.append(f"üìä Overall system success rate: {success_rate:.1f}% of claims are completed successfully")
    
    # Food type insights
    if "Query 12: Meal Type Popularity" in results:
        meal_data = results["Query 12: Meal Type Popularity"]
        top_meal = meal_data.iloc[0]
        insights.append(f"üçΩÔ∏è  {top_meal['meal_type']} is the most popular meal type with {top_meal['total_quantity_claimed']:,.0f} units claimed")
    
    # Utilization insights
    if "Query 7: Common Food Types" in results:
        food_type_data = results["Query 7: Common Food Types"]
        avg_utilization = food_type_data['percentage_of_total_quantity'].mean()
        insights.append(f"‚ôªÔ∏è  Average food utilization across types: {avg_utilization:.1f}%")
    
    return insights

insights = generate_insights()

print("üìà KEY FINDINGS:")
for i, insight in enumerate(insights, 1):
    print(f"{i}. {insight}")

print(f"\nüí° RECOMMENDATIONS:")
recommendations = [
    "üéØ Focus on expanding successful provider types to increase food supply",
    "üìç Replicate successful distribution models from top-performing cities",
    "‚ö° Improve claim processing efficiency to reduce cancellation rates",
    "ü§ù Enhance provider-receiver matching in underperforming areas",
    "üìä Implement real-time monitoring for expiry date management",
    "üîÑ Develop feedback loops between providers and receivers",
    "üì± Consider mobile app for easier claim management",
    "üìà Set up automated alerts for high-demand food types"
]

for i, rec in enumerate(recommendations, 1):
    print(f"{i}. {rec}")

print(f"\nüéâ Analysis completed! Database contains {len(results)} comprehensive queries with actionable insights.")
print("Ready for Streamlit dashboard development!")
```
